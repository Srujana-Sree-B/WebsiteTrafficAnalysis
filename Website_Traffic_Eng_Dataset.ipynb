```python
import pandas as pd
```

```python
my_data = pd.read_csv("C:/Users/sruja/Downloads/Dataset/website_wata.csv")
```

```python
my_data = my_data.drop_duplicates()
my_data
```

```python
# Check for Missing Values
print("Missing Values in Each Column:")
print(my_data.isnull().sum())
```

```python
# Our dataset doesnot have any missing values. But lets assume it has some missing values.
# In that case we need to replace missing values with mean of specific column.
```

```python
for column in my_data.select_dtypes(include=['number']).columns:  # Only process numeric columns
    if my_data[column].isnull().sum() > 0:  # Check if the column has missing values
        my_data[column] = my_data[column].fillna(my_data[column].mean())
```

```python
my_data
```

```python
#Standardize the url, if exist anywher in our dataset
if 'url' in my_data.columns:
    my_data['url'] = my_data['url'].str.lower()
else:
    print("Column 'url' does not exist in the dataset.")
```

```python
# Time spent per page: Average time spent per page view
my_data['Time Spent Per Page'] = my_data['Time on Page'] / my_data['Page Views']
```

```python
# Session Duration Per Visit: Total session duration divided by previous visits
# Adding 1 to avoid division by zero if there are no previous visits
my_data['Session Duration Per Visit'] = my_data['Session Duration'] / (my_data['Previous Visits'] + 1)
```

```python
# Categorize traffic sources into a numerical values
traffic_mapping = {'Direct': 1, 'Referral': 2, 'Social Media': 3, 'Organic': 4}
my_data['Traffic Source Code'] = my_data['Traffic Source'].map(traffic_mapping)
```

```python
# To give an insight into how long users are staying per page.
my_data['Session Length to Page Views Ratio'] = my_data['Session Duration'] / my_data['Page Views']
```

```python
my_data
```

```python
# Identifying the top pages based on page views (this assumes each row represents a page)
top_pages = my_data['Page Views'].value_counts().head(5)
print(f"Top pages based on traffic:\n{top_pages}")
```

```python
# Identifying low-traffic pages (pages with fewer views)
low_traffic_pages = my_data['Page Views'].value_counts().tail(5)
print(f"Low-traffic pages:\n{low_traffic_pages}")
```

```python
# Flag frequent visitors (Previous Visits > 3)
my_data['Frequent Visitor'] = my_data['Previous Visits'] > 3
print(f"Frequent Visitors:\n{my_data[['Previous Visits', 'Frequent Visitor']]}")
```

```python
# Flag first-time users (Previous Visits = 0)
my_data['First-time User'] = my_data['Previous Visits'] == 0
print(f"First-time Users:\n{my_data[['Previous Visits', 'First-time User']]}")
```

```python
# Flag high converting users (Conversion Rate > 0.9)
my_data['High Converting User'] = my_data['Conversion Rate'] > 0.9
print(f"High Converting Users:\n{my_data[['Conversion Rate', 'High Converting User']]}")
```
